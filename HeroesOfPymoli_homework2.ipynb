```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
player_count=len(purchase_data["SN"].unique())

player_df=pd.DataFrame({"Total Players":[player_count]})

player_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(purchase_data["Item Name"].unique())

num_purchases=len(purchase_data["Purchase ID"])

avg_price=(purchase_data['Price']).mean()

total_price=purchase_data['Price'].sum()

purchase_analysis=pd.DataFrame({"Number of Unique Items":[unique_items],
                                "Average Price":pd.Series([avg_price]).map("${:,.2f}".format),
                                "Number of Purchases":[num_purchases], 
                                "Total Revenue":pd.Series([total_price]).map("${:,.2f}".format)})

purchase_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_df=purchase_data.loc[:,["SN","Gender"]]
gender_no_dups=gender_df.drop_duplicates(subset='SN')

new_gender=gender_no_dups.groupby("Gender").count()


gender_count=gender_no_dups.groupby("Gender").count()/player_count*100

gender_count["Total Count"]=new_gender

gender_count=gender_count.rename(columns={"SN":"Percentage of Players"})

gender_count=gender_count[["Total Count", "Percentage of Players"]]

gender_count
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
PA_gender_count=purchase_data.groupby("Gender").count()

PA_gender["Purchase Count"]=PA_gender_count["Purchase ID"]
PA_gender["Average Purchase Price"]=purchase_data.groupby("Gender").mean()["Price"]
PA_gender["Total Purchase Value"]=purchase_data.groupby("Gender").sum()["Price"]
PA_gender["Avg Total Purchase per Person"]=purchase_data.groupby("Gender").sum()["Price"]/new_gender["SN"]

PA_gender_df=PA_gender[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]]

PA_gender_df


# Pandas didn't like what I was trying to do with the formatting. Not sure what was wrong.

# PA_df=PA_gender["Purchase Count"]
# PD_df["Average Purchase Price"]=PA_gender["Average Purchase Price"].map("${:,.2f}".format).values
# PA_df["Total Purchase Value"]=PA_gender["Total Purchase Value"].map("${:,.2f}".format).values
# PA_df["Avg Total Purchase per Person"]=PA_gender["Avg Total Purchase per Person"].map("${:,.2f}".format).values

# PA_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
custom_bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 100]
bin_labels=["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

no_dups=purchase_data.drop_duplicates(subset=['SN'])
no_dups["Age Bins"]=pd.cut(no_dups["Age"], bins=custom_bins, labels=bin_labels)

age_bins=no_dups.groupby("Age Bins").count()

age_bins["Percent of Players"]=((age_bins["SN"]/player_count)*100).map("{:,.2f}%".format).values

bins_and_percent=age_bins[["Purchase ID", "Percent of Players"]]

bins_and_percent
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
custom_bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 100]
bin_labels=["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data["Age Bins"]=pd.cut(purchase_data["Age"], bins=custom_bins, labels=bin_labels)


age_df=purchase_data.loc[:,["SN","Age Bins"]]
age_no_dups=age_df.drop_duplicates(subset='SN')
new_age=age_no_dups.groupby("Age Bins").count()


PA_age["Purchase Count"]=purchase_data.groupby("Age Bins").count()  #purchase count column totals
PA_age["Average Purchase Price"]=purchase_data.groupby("Age Bins").mean()["Price"] #avg purchase price
PA_age["Total Purchase Value"]=purchase_data.groupby("Age Bins").sum("Price")  #total purchase value column
PA_age["Avg Total Purchase per Person"]=purchase_data.groupby("Age Bins").sum()["Price"]/new_age["SN"] #avg total PPP

PA_age_df=PA_age[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]]

PA_age_df

# I have this set up the same way as the Purchasing Analysis (Gender) section yet it won't create a dataframe like the gender code does.
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_count=purchase_data[["SN", "Purchase ID", "Price"]]

sn_count=purchase_count[["SN", "Price"]].groupby("SN").count()


total_purch_val=purchase_count[["SN", "Price"]].groupby("SN").sum()

avg_purch_price=purchase_count[["SN", "Price"]].groupby("SN").mean()


sn_count["Average Purchase Price"]=avg_purch_price
sn_count["Total Purchase Value"]=total_purch_val
sn_count=sn_count.rename(columns={"Price":"Purchase Count"})

sn_count.sort_values(by='Total Purchase Value', ascending=False).head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_data=purchase_data[["Item ID", "Item Name", "Price"]]

item_grouped=item_data.groupby(["Item ID", "Item Name"]).sum()

item_grouped["Purchase Count"]=item_data.groupby(["Item ID", "Item Name"]).count()

item_grouped=item_grouped.rename(columns={"Price":"Total Purchase Value"})

#item_grouped["Price New"]=pd.merge(item_grouped,purchase_data[['Price','Price New']],on='Price', how='right')   (not working)


item_grouped.sort_values(by='Purchase Count', ascending=False).head()
```
